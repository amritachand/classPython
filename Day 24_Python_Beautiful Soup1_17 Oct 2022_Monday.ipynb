```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
soup = requests.get(url)
code = soup.text

lxml_code= BeautifulSoup(code, "lxml")
lxml_code
```

```python
table_code = lxml_code.table
table_code
```

```python
table_code = lxml_code.table
tags = table_code.find_all('tr')
tags
```

```python
# using for loop

for tag in tags:
    print(tag.text)
```

```python
raw_data = []
for tag in tags:
    raw_data.append(tag.text)
    
raw_data
```

```python
raw_data = []
for tag in tags:
    x = tag.text.split('\n')
    raw_data.append(x)
    
raw_data
```

```python
import csv
file=open ('covid.csv','w')
x = csv.writer(file)
x.writerows(raw_data)
```
